По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

## Описание данных
 - airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
 - balcony — число балконов
 - ceiling_height — высота потолков (м)
 - cityCenters_nearest — расстояние до центра города (м)
 - days_exposition — сколько дней было размещено объявление (от публикации до снятия)
 - first_day_exposition — дата публикации
 - floor — этаж
 - floors_total — всего этажей в доме
 - is_apartment — апартаменты (булев тип)
 - kitchen_area — площадь кухни в квадратных метрах (м²)
 - last_price — цена на момент снятия с публикации
 - living_area — жилая площадь в квадратных метрах(м²)
 - locality_name — название населённого пункта
 - open_plan — свободная планировка (булев тип)
 - parks_around3000 — число парков в радиусе 3 км
 - parks_nearest — расстояние до ближайшего парка (м)
 - ponds_around3000 — число водоёмов в радиусе 3 км
 - ponds_nearest — расстояние до ближайшего водоёма (м)
 - rooms — число комнат
 - studio — квартира-студия (булев тип)
 - total_area — площадь квартиры в квадратных метрах (м²)
 - total_images — число фотографий квартиры в объявлении
 
Пояснение: апартаменты — это нежилые помещения, не относящиеся к жилому фонду, но имеющие необходимые условия для проживания.

### Обработка пропусков

```python
#Замена тип данных на timedate

df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'])

cols_date = ['first_day_exposition']

#Замена тип данных на boolean
df['is_apartment'] = df['is_apartment'].astype(bool)

#Список стольбцов с типом данных float64
cols_float = list(df.select_dtypes(include = ['float64']).columns)

#Список стольбцов с типом данных int64
cols_int = list(df.select_dtypes(include = ['int64']).columns)

#Список стольбцов с типом данных boоlean
cols_bool = list(df.select_dtypes(include = ['bool']).columns)
               
#Список стольбцов с типом данных object

cols_obj = [x for x in df.columns if x not in cols_float + cols_int + cols_bool + cols_date]

print(cols_float, '\n\n', cols_int, '\n\n',cols_bool, '\n\n', cols_obj)
```

```python

#Замена пропушеных значений типа float64 на медиану
for col in cols_float:
    df[col] = df[col].fillna(value = df[col].median())
    
#Замена пропушеных значений типа float64 на медиану
for col in cols_int:
    df[col] = df[col].fillna(value = df[col].median())
    

#Замена пропушеных значений типа object  на 'не указано'
for col in cols_obj:
    df [col] = df[col].fillna(value = 'не указано')

    
print(df.isna().sum())
```

### Замена типа данных

```python
#Замена тип float на int признаки которые не могут физический быить бробными

cols_not_part = ['floors_total','balcony','days_exposition', 'parks_around3000' , 'ponds_around3000',
                  'total_area', 'last_price']

for cols in cols_not_part:
    df[cols] = df[cols].astype('int64')


print(df.info())
```

### Шаг 3. Посчитайте и добавьте в таблицу

```python
#Посчитаем ценна за м2
df['price_per_m2'] = ( df['last_price'] / (df['total_area']*1000) ).astype('int')

df['date_publication'] = df['first_day_exposition']

# Разделяем признак дата публикаций
df['publication_year'] = df['date_publication'].dt.year
df['publication_day'] = df['date_publication'].dt.weekday
df['publication_month'] = df['date_publication'].dt.month

#Функция групировки по этажам
def floor_group (row):
    floor = row['floor']
    floors_total = row['floors_total']
    
    if floor == 1:
        return 'Первый'
    if floor == floors_total:
        return 'Последный'
    else:
        return 'Другой'
# Добавляем признак тип этажа    
df['floor_group'] = df.apply(floor_group, axis =1 )
#Проверяем количество уникальгых значений
print(df['floor_group'].value_counts())

#Считаем доля жилого площади
df['percent_living_total_area'] = ((df['living_area'] / df['total_area'])*100).astype('int')

#Считаем доля  площади кухний
df['percent_kitchen_total_area'] = ((df['kitchen_area'] / df['total_area'])*100).astype('int')

#Считаем доля  ростояние до центра в км
df['distance_to_cityCenter_km'] = (df['cityCenters_nearest'] / 1000).astype('int')

print()
print(df.head(2).T)
  
```
