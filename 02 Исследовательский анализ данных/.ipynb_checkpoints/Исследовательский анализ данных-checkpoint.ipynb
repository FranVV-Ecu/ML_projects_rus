### Обработка пропусков

```python
df['ceiling_height'] =  (
                        df['ceiling_height']
                        .fillna(value = 2.7)
)


df['floors_total'] =   (
                         df['floors_total']
                        .fillna(value = df['floors_total'].median())
)

df['living_area'] = (
                        df['living_area']
                        .fillna(value = df['living_area'].median())
)

df['is_apartment'] = (
                        df['is_apartment']
                        .fillna(value = 'Не указано')
)

df['kitchen_area'] = (
                        df['kitchen_area']
                        .fillna(value = df['kitchen_area'].median())
)

df['balcony'] = (
                        df['balcony']
                        .fillna(value = 0 )
)

df['locality_name'] = (
                        df['locality_name']
                        .fillna(value = 'Не указано')
)

df['airports_nearest'] = (
                        df['airports_nearest']
                        .fillna(value = df['airports_nearest'].median())
)

df['cityCenters_nearest'] = (
                        df['cityCenters_nearest']
                        .fillna(value = df['cityCenters_nearest'].median())
)

df['parks_around3000'] = (
                        df['parks_around3000']
                        .fillna(value = df['parks_around3000'].median())
)

df['parks_nearest'] = (
                        df['parks_nearest']
                        .fillna(value = df['parks_nearest'].median())
)

df['ponds_around3000'] = (
                        df['ponds_around3000']
                        .fillna(value = df['ponds_around3000'].median())
)

df['ponds_nearest'] = (
                        df['ponds_nearest']
                        .fillna(value = df['ponds_nearest'].median())
)

df['days_exposition'] = (
                        df['days_exposition']
                        .fillna(value = df['days_exposition'].median())
)

print(df.isna().sum())
```

### Замена типа данных

```python
df['floors_total'] = df['floors_total'].astype('int')
df['balcony'] = df['balcony'].astype('int')
df['days_exposition'] = df['days_exposition'].astype('int')
df['parks_around3000'] = df['parks_around3000'].astype('int')
df['ponds_around3000'] = df['ponds_around3000'].astype('int')
df['days_exposition'] = df['days_exposition'].astype('int')
df['days_exposition'] = df['days_exposition'].astype('int')
df['total_area'] = df['total_area'].astype('int')
df['last_price'] = df['last_price'].astype('int')

print(df.info())
```

### Шаг 3. Посчитайте и добавьте в таблицу

```python
df['price_per_m2'] = ( df['last_price'] / (df['total_area']*1000) ).astype('int')

df['date_publication'] = pd.to_datetime(df['first_day_exposition'], format = '%Y-%m-%d')

df['publication_year'] = df['date_publication'].dt.year
df['publication_day'] = df['date_publication'].dt.weekday
df['publication_month'] = df['date_publication'].dt.month

def floor_group (row):
    floor = row['floor']
    floors_total = row['floors_total']
    
    if floor == 1:
        return 'Первый'
    if floor == floors_total:
        return 'Последный'
    else:
        return 'Другой'
    
df['floor_group'] = df.apply(floor_group, axis =1 )
print(df['floor_group'].value_counts())

df['percent_living_total_area'] = ((df['living_area'] / df['total_area'])*100).astype('int')

df['percent_kitchen_total_area'] = ((df['kitchen_area'] / df['total_area'])*100).astype('int')

df['distance_to_cityCenter_km'] = (df['cityCenters_nearest'] / 1000).astype('int')

print()
print(df.head(5))
  
```

### Вывод

Расчитал и добавлял в таблицу  стоимость квадратного метра, день, месяц, год ge,публикации, этаж квартиры, порцентное соотношение жилой площади к общей и порцентное соотношение площади кухни к общей площади. Все результаты перевели в целовой тип данных.

Средная стоимость квадратного метра в центре высше чем за ее пределами.

При увеличение высоты потолков стоимости недвижимости тоже увеличивается.

Тенденции стоимость по отношению к году, месяца и дня снятия из публикации сохраянияются как и для всей недвижимости по городу.
