## Расчет для каждого пользователя

Расчет для каждого пользователя:
количество сделанных звонков и израсходованных минут разговора по месяцам;
количество отправленных сообщений по месяцам;
объем израсходованного интернет-трафика по месяцам;
помесячную выручку с каждого пользователя


```python
# Добавление столбцы с указанием на месяц
data_messages['month'] = data_messages['message_date'].dt.month
calls['month'] = calls['call_date'].dt.month

```

```python
# Соединяем таблицу с данными о клиентах и звонков.
pivot_users = data_users[['user_id','tariff', 'city']]
data_merge =  calls.merge(pivot_users, on='user_id', how='left')
data_merge.head()
```

```python
# Создаем свобные таблицы звонков длительности больше чем 0 с по пользователям и месяцам. Выводим счет и сумму значений.
pivot_calls = calls.pivot_table(index = ['user_id','month'], values = 'rounded_calls' , aggfunc= ['count','sum'])
pivot_calls.head()
```

```python
# Создаем свобные таблицы сообщений  по пользователям и месяцам. Выводим счет.
pivot_messages = data_messages.pivot_table(index = ['user_id','month'], values = 'id', aggfunc= ['count'])
pivot_messages.head()
```

```python
# Создаем свобные таблицы трафика интернета по пользователям и месяцам. Выводим сумму трафика.
pivot_internet = data_internet.pivot_table(index = ['user_id','month'], values = 'gb_used', aggfunc= ['sum'])
pivot_internet.head()
```

```python
# Соединиаем полученные данные сгрупированые по месяцам в один датасет.
# Применяем метод merge к датасетам сводным таблицам ранее подготовленые
usage_merge = pivot_calls.merge(pivot_messages, on =['user_id','month'] , how = 'left')

sample_data = usage_merge.merge(pivot_internet,  on = ['user_id','month'], how = 'left')


# Переменуем столбцы датасета
sample_data.columns = ['call_count', 'call_minutes', 'sms_count', 'gb_used']

sample_data.head()
```

```python
#Получаем датасет для расчета выручки клиентов по месяцам и тарифам кажлого пользователя
full_data = sample_data.join(pivot_users.set_index('user_id'), on = ['user_id'] , how='left').reset_index().fillna(0)

full_data.head(10)


```

```python
#Функция расчета помесячную вырочку по звонкам
def revenue_calls (row):
    tariff = row['tariff']
    if tariff == 'smart':
        extra_minutes = row['call_minutes'] - 500
        if extra_minutes > 0 :
            call_cost =  550 + extra_minutes*3
        else :
            call_cost = 550
    elif tariff == 'ultra':
        extra_minutes = row['call_minutes'] - 3000
        if extra_minutes > 0 :
            call_cost =  1950 + extra_minutes*1
        else :
            call_cost = 1950
    return call_cost

#Функция расчета помесячную вырочку по сообщениям
def revenue_sms (row):
    tariff = row['tariff']
    if tariff == 'smart':
        extra_sms = row['sms_count'] - 50
        if extra_sms > 0 :
            cost_sms =  extra_sms*3
        else :
            cost_sms = 0
    elif tariff == 'ultra':
        extra_sms = row['sms_count'] - 1000
        if extra_sms > 0 :
            cost_sms =  extra_sms*1
        else :
            cost_sms = 0
    return cost_sms

#Функция расчета помесячную вырочку по Гб
def revenue_gb (row):
    tariff = row['tariff']
    if tariff == 'smart':
        extra_gb = row['gb_used'] - 15
        if extra_gb > 0 :
            cost_gb =  (extra_gb)*200
        else :
            cost_gb = 0
    elif tariff == 'ultra':
        extra_gb = row['gb_used'] - 30
        if extra_gb > 0 :
            cost_gb =  (extra_gb)*150
        else :
            cost_gb = 0
    return cost_gb
```

<font color="blue">использование констант не очень хорошое дело. Представь что потом поменяют числа в датасете, лезть в код? Ну такое. Плюсом строго вбиты 2 тарифа. Конечно это больше кодерское, но всё же

```python
#  Складиваем помесячную выручку пользователей по звонкам, соббщений и трафика интернета
full_data['revenue_month'] = ( full_data.apply(revenue_calls, axis = 1)+

 full_data.apply(revenue_sms, axis = 1) +

 full_data.apply(revenue_gb, axis = 1) )


full_data.head(10)
```

```python
#Группировка данных по тарифам и месяцам с  расчетом средное значение
stat_data = full_data.groupby(['tariff', 'month']).mean().round(2)

stat_data.columns = ['user_id', 'count_calls', 'calls', 'sms', 'gb', 'revenue']

stat_data.head()
```

```python
# Группировка по тарифу смарт
smart = stat_data.query('tariff == "smart"')

# Расчет среднего значения 
print("Средные значения тарифа Смарта")

for column in smart:
    mean = smart[column].mean().round(2)
    print(column,':', mean )
```

```python
# Расчет дисперсии тарифа смарта
print("Дисперсии тарифа Смарта")
for column in smart:
    variance = np.var(smart[column], ddof= 1).round(2)
    print(column, ':' , variance)
```

```python
# Расчет стандартного отклонения тарифа смарта

print("Стандартное отклонения значений тарифа Смарта")
for column in smart:
    standart_deviation = np.std(smart[column], ddof= 1).round(2)
    print(column, ':' , standart_deviation)
```

```python
# Группировка по тарифу смарт
ultra = stat_data.query('tariff == "ultra"')

# Расчет среднего значения 
print("Средные значения тарифа Ультра")
for column in stat_data:
    mean = ultra[column].mean().round(2)
    print(column,':', mean )
```

```python
# Расчет дисперсии тарифа ультра
print("Дисперсии тарифа Ультра")
for column in ultra:
    variance = np.var(ultra[column], ddof= 1).round(2)
    print(column, ':' , variance)
```

```python
# Расчет стандартного отклонения тарифа смарта

print("Стандартное отклонения значений тарифа ультра")
for column in ultra:
    standart_deviation = np.std(ultra[column], ddof= 1).round(2)
    print(column, ':' , standart_deviation)
```
