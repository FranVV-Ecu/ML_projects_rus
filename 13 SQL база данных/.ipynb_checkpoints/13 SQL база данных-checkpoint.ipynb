```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

#Определение функций вывода графика горизнтальной cтолбчатой диаграммы
def plot_bars_horizontal(x_data, y_labels,lenght, wide, x_label, title):
    fig, ax = plt.subplots(figsize=(lenght, wide ))
    plt.barh(y_labels, x_data, align='center', alpha=0.6, color ='brgcmky')
    plt.yticks(y_labels)
    plt.xlabel(x_label)
    plt.title(title)
    plt.grid()
    ax.invert_yaxis()
    plt.show()

# Вывод графика количетсва полетов по моделям самолётов
plot_bars_horizontal(f_amount.flights_amount,f_amount.model, 8, 4 , 'Flights amount' , 'Flights amount by city at september')

#Определение область круговой диаграммы
fig, ax = plt.subplots(figsize=(10, 7), subplot_kw=dict(aspect="equal"))

#Определение лабелы и данных для их построение
model = f_amount.model
data = f_amount.flights_amount
#
models = [x.split()[-1] for x in model]
explode = np.full((1,len(f_amount.flights_amount)), 0.05)

#Функция расчета порцентного соотношения
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, explode = explode[0], autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"), shadow=True, startangle=90)

ax.legend(wedges, model,
          title="Aircraft models",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Ratio amount fligts by aircraft model")

plt.show()
```

### Вывод
Первый три места ханимают следющие модели соответсвено: Cessana 208 Caravan, Bombardier CRJ-20 и Sukhoi Superjet -100.
По количество полётов первые три метса с отрывом отличаются от осталных. Их суммарное порцентное соотношение составляет около 80% от общего количетсво полётов. Эти модели являются не большими самолётами. Далле рассмотрим мощглсть перевоза пассажиров.

```python
# Пассажировместимость каждой модели. Информация взата из откритых источников (Wikipedia)
passangers =  {'Cessna 208 Caravan': 13 ,
              'Bombardier CRJ-200': 50,
              'Sukhoi SuperJet-100': 87,
              'Airbus A321-200': 195   ,
              'Boeing 737-300':   149  ,
              'Airbus A319-100':  136 ,
              'Boeing 767-300':    316 ,
              'Boeing 777-300':   409 }
# Добавляем стольбец Пассажировместимость
f_amount['capacity_passangers'] = f_amount['model'].map(passangers)

# Добавляем стольбец мощность пассажировместимость
f_amount['max_passangers'] = f_amount['flights_amount']*f_amount['capacity_passangers']

#Сортировка по  мощности пассажировместимость
f_amount = f_amount.sort_values(by='max_passangers', ascending =  False)

f_amount
```

```python
# Вывод графика мощность пассажировместимост в по моделям самолётов
plot_bars_horizontal(f_amount.max_passangers,f_amount.model, 8, 4 , 'Max passangers amount' , 'Max passangers amount by city at september')

#Определение область круговой диаграммы
fig, ax = plt.subplots(figsize=(10, 7), subplot_kw=dict(aspect="equal"))

#Определение лабелы и данных для их построение
model = f_amount.model
data_2 = f_amount.max_passangers

models = [x.split()[-1] for x in model]
explode = np.full((1,len(f_amount.max_passangers)), 0.05)


wedges, texts, autotexts = ax.pie(data_2, explode = explode[0], autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"), shadow=True, startangle=90)

ax.legend(wedges, model,
          title="Aircraft models",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Ratio amount max passangers by aircraft model")

plt.show()
```

### Вывод

При расмотрении мощность пассажировместимость по моделям рейтинг изменился по сравнению с рейтингом по количество полётов. Первый четырые местя занимают : Sukhoi Superjet -100, Bombardier CRJ-200, Boeing 767-300 и Airbus A321-200 соответсвенно. Их суммарное порцентное соотношение составляет 72,9 %. # 2. Проверка гипотезы

```python
# Зашрузка датасет количества билетов по неделям и наличие фестивалов.
f_demand = pd.read_csv('/datasets//query_last.csv')
f_demand
```

```python
# Выявление индеки недели без фестивалей
index = f_demand.festival_name.isna()

# Массив
no_festivals_demand = f_demand.ticket_amount[index].reset_index(drop=True)

festivals_demand = f_demand.ticket_amount[~index].reset_index(drop=True)

print('Week festival  demand:\n', festivals_demand, '\n','\n',
      'Week no festival demand:\n', no_festivals_demand)
```

## Постановка задачи
Гипотеза: «Средний спрос на билеты во время фестивалей не отличается от среднего спроса на билеты в обычное время».

Для проверки гипотезы испольщуем критерий  о равенстве средних двух генеральных совокупностей, потому что проверяем совокупности количества билетов всех городов, то есть средной спроса по неделям.

Нулевая гипотеза - Спрос на билеты во время фестивалей равно  спросу на билеты в обычное время

Альтернативная гипотеза - Спрос на билеты во время фестивалей не равно  спросу на билеты в обычное время

```python
from scipy import stats as st
alpha =0.05
# Проверим гипотезу о равенстве средних двух генеральных совокупностей
results = st.ttest_ind(
    festivals_demand, 
   no_festivals_demand,
  equal_var= True)

# выведим на экран получившееся p-значени
print('p-значение: ', results.pvalue)

#  сравнение  получившееся p-значение с заданным уровнем статистической значимости):
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

## Вывод

Нет достаточно сведений для того чтобы опробергнуть нулевую гипотезу. То есть, спрос на билеты во время фестивалей равно  спросу на билеты в обычное время.

Скорее всего фестивали ни вляют на спрос билетов.
